```python
import pandas as pd
import numpy as np
```

```python
df1_similarity = pd.read_csv("../Dataset/small/similarity-output/part-00000", \
                             names = ["user_x", "user_y", "similarity_score"])
```

```python
df2_similarity = pd.read_csv("../Dataset/small/similarity-output/part-00001", \
                             names = ["user_x", "user_y", "similarity_score"])
```

```python
df2_similarity.head()
```

```python
user_user_similarity = df1_similarity.append(df2_similarity,ignore_index=True)
```

```python
user_user_similarity.shape
```

```python
user_user_similarity.head()
```

```python
user_repo_rating =  pd.read_csv("../Dataset/small/user_repo_ratings_small.csv")
```

```python
user_repo_rating.head()
```

```python
dummy_user_id =  135
```

```python
######### Find Similar Users ############################
```

```python
similar_users_data = user_user_similarity[ (user_user_similarity.user_x == dummy_user_id) & \
                                          (user_user_similarity.similarity_score != 0)]. \
                    sort(["similarity_score"], ascending = False)
```

```python
print "Found %d other users similar to user_id %d "%(similar_users_data.shape[0],dummy_user_id)
```

```python
similar_users_data.head()
```

```python
similar_users_repos = user_repo_rating[ user_repo_rating.index_x.isin(similar_users_data.user_y)]
```

```python
print "Found total %d repos of similar users "%(similar_users_repos.shape[0])
```

```python
# similar_users_repos["total_rating"] = similar_users_repos.loc[:,('bias_score')] + similar_users_repos.loc[:,('rating')]
```

```python
top_percentile = np.percentile(similar_users_repos.rating,96)  
```

```python
top_percentile
```

```python
similar_users_repos = similar_users_repos[ similar_users_repos.rating >= top_percentile ]
```

```python
similar_users_repos.head()
```

```python
similar_users_repos.sort(["bias_score"], ascending = False, inplace=True)
```
